```python
df = pd.read_excel('/content/drive/MyDrive/DataViz23Data/ectreport66.xlsx', sheet_name='Candidate_Constituency')
df1 = pd.read_excel('/content/drive/MyDrive/DataViz23Data/ectreport66.xlsx',sheet_name='result_constituencies_Candidate')
df2 = pd.read_excel('/content/drive/MyDrive/DataViz23Data/ectreport66.xlsx', sheet_name='info_party_overview')
```

```python
ect = pd.concat([df, df1, df2], axis=1, join="inner")
```

```python
ect
```

```python
ect1 = ect[['mp_app_name','name','mp_app_vote']]
ect1
```

```python
ect00 = ect1.groupby('mp_app_name')['mp_app_vote'].max().reset_index().sort_values(by='mp_app_vote',ascending=False).head(10)
```
