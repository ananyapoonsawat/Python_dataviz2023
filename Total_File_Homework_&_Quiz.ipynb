ข้อที่ 1 12:30:15 - 13:41:07

```python
q1 = '12:30:15'
Q1 = q1.split(":")
hours1 = int(Q1[0])*3600
minutes1 = int(Q1[1])*60
seconds1 = int(Q1[2])

q2 = '13:41:07'
Q2 = q2.split(":")
hours2 = int(Q2[0])*3600
minutes2 = int(Q2[1])*60
seconds2 = int(Q2[2])

a1 =  hours1 + minutes1 + seconds1
a2 =  hours2 + minutes2 + seconds2
a3 = a2 - a1

ho = int(a3/3600)
mi = int((a3%3600)/60)
se = int(a3%60)

print(f' ตั้งเเต่ 12:30:15 - 13:41:07 จะต่างกันเป็นเวลา {ho}ชั่วโมง {mi}นาที {se}วินาที')
```

```python
import numpy as np

def print_kaokai(numbg = 0):
  int_bg = int(numbg)
  num_logo = 99
  MFP_17 = int_bg*np.ones((17,17))

  #outter 7
  #ขอบบน ---
  MFP_17[0:1, 0:17] = num_logo
  MFP_17[1:2, 1:16] = num_logo

  # /
  MFP_17[1:3, 14:16] = num_logo
  MFP_17[3:5, 13:15] = num_logo
  MFP_17[5:7, 12:14] = num_logo
  MFP_17[7:9, 11:13] = num_logo
  MFP_17[9:11, 10:12] = num_logo
  MFP_17[11:13, 9:11] = num_logo
  MFP_17[13:15, 8:10] = num_logo

  # เก็บหาง
  MFP_17[14:15, 7:10] = num_logo
  MFP_17[15:17, 8:9] = num_logo


  #inner 7
  #ขอบบน ---
  MFP_17[4:5, 2:11] = num_logo
  MFP_17[5:6, 3:10] = num_logo

  # /
  MFP_17[5:7, 8:10] = num_logo
  MFP_17[7:9, 7:9] = num_logo
  MFP_17[9:11, 6:8] = num_logo

  #เก็บหาง
  MFP_17[10:11, 5:8] = num_logo
  MFP_17[11:13, 6:7] = num_logo
  print(MFP_17)

print_kaokai(5)
```
